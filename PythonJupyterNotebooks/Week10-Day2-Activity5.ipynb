```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df['NumberPassengers'], order=(2, 1, 2))
```

```python
results = model.fit()
results.summary()
```

```python
pd.DataFrame(results.forecast(steps=3)[0]).plot(title="Passenger Travel Forecast")
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)



train_size = 0.75
train_df = df[0:int(len(df)*train_size)]


def get_predictions(train_df, p, d, q, num_steps=20): 
    model = ARIMA(train_df , order=(p,d,q))
    results = model.fit()
    print(results.summary())
    test_predictions = pd.DataFrame(results.forecast(steps=num_steps)[0])
    test_predictions.columns = ['NumberPassengers']

    start_index = int(len(df)*train_size)
    end_index = start_index + num_steps

    test_predictions.index = df[start_index:end_index].index
    return test_predictions 



combinations_to_test = [(2,1,2), (1,1,1), (2,1,1), (1,1,2), (2,1,4)]

summary = []
for combination in combinations_to_test: 
    try: 
        p = combination[0]
        d = combination[1]
        q = combination[2]
        print("Adding output from model p:{}, d:{}, q:{}".format(p,d,q))
        output = get_predictions(train_df, p, d, q, num_steps=20)
        model_name = 'model p:{}, d:{}, q:{}'.format(p,d,q)
        summary.append([output, model_name])
        
    except: 
        print("p:{}, d:{}, q:{} are not viable".format(p,d,q))
        
        
import matplotlib.pyplot as plt
num_steps= 20
true_df = df.iloc[0:len(train_df)+num_steps]


plt.figure(figsize=(20,10))
plt.plot(true_df.index, true_df['NumberPassengers'], label='Original Signal')

for model in summary: 
    test_predictions = model[0]
    model_name = model[1]
    plt.plot(test_predictions.index, test_predictions['NumberPassengers'],label=model_name)
    
plt.title("Comparing True values with Predicted Values", fontsize=20)
plt.ylabel("Number of Passengers")
plt.xlabel("Date")
plt.legend()
```
