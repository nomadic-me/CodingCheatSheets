```python
import pandas as pd
df = pd.read_csv("Resources/Week10-Day1-Activity1-liquor_sales.csv",
                parse_dates=['datetime'],
                index_col=['datetime'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df.plot()
```

```python
decompose_df = seasonal_decompose(df['liquor_sales'], 
                                 model='additive')
decompose_df.plot()
```

```python
decompose_df = seasonal_decompose(df['liquor_sales'], 
                                 model='multiplicative')
decompose_df.plot()
```
